```python
import pandas as pd
import re

df = pd.read_csv("../data/raw/deceptive-opinion.csv")
df.head()
```

```python
print("Önce:", df.shape)
print("Duplicate text sayısı:", df["text"].duplicated().sum())

df = df.drop_duplicates(subset=["text"]).reset_index(drop=True)

print("Sonra:", df.shape)
print("Duplicate text sayısı:", df["text"].duplicated().sum())
```

```python
df = df[["text", "deceptive"]].copy()

print("Yeni shape:", df.shape)
df.head()
```

```python
df["label"] = df["deceptive"].map({
    "truthful": 0,
    "deceptive": 1
})

print(df[["deceptive", "label"]].head())
print("\nLabel dağılımı:")
print(df["label"].value_counts())
```

```python
def clean_text(text):
    text = text.lower()                       # küçük harfe çevir
    text = re.sub(r"[^a-zA-Z\s]", " ", text) # harf ve boşluk dışını temizle
    text = re.sub(r"\s+", " ", text).strip() # fazla boşlukları temizle
    return text
```

```python
df["clean_text"] = df["text"].apply(clean_text)

df[["text", "clean_text", "label"]].head()
```

```python
print("Final shape:", df.shape)
print("\nEksik değerler:")
print(df.isnull().sum())
```

```python
df.to_csv("../data/processed/clean_reviews.csv", index=False)
print("Kaydedildi: ../data/processed/clean_reviews.csv")
```
